```python
def fetch_live_price(ticker_symbol):
    """
    Fetches the live market price for a given ticker symbol using yfinance.
    Returns the price if successful, None otherwise.
    """
    try:
        ticker = yf.Ticker(ticker_symbol)
        # Get the most recent closing price
        current_price = ticker.history(period="1d")['Close'][0]
        return current_price
    except Exception as e:
        print(f"Could not fetch live price for {ticker_symbol}: {e}")
        return None
```

```python
class FinancialAnalyzer:
    def __init__(self):
        self.currency = ""
        self.currency_symbol = ""
        self.currency_map = {"india": "₹", "usa": "$"}
        self.portfolio = {}
        self.goals = {}
        self.risk_tolerance = ""
        self.financial_situation = {"debts": 0, "expenses": 0, "salary": 0, "other_income": {"interest": 0, "rent": 0}, "annual_income": 0} # Added other_income and annual_income
        self.asset_allocation = {}
        self.retirement_goals = {}
        self.emergency_fund = {"amount": 0, "location": ""}
        self.future_debt_planning = ""
        self.tax_situation = {"type": "", "regime": "", "implications": ""}
        self.insurance_coverage = ""
        self.interest_in_esg = ""
        self.other_factors = {}
        self.stock_details = {}
        self.bond_details = {}
        self.other_investments_details = {}


    def collect_user_inputs(self):
        # Currency selection
        while self.currency.lower() not in ["india", "usa"]:
            self.currency = input("Are you from India or USA? ").lower()
            if self.currency in self.currency_map:
                self.currency_symbol = self.currency_map[self.currency]
            else:
                print("Invalid country. Please enter India or USA.")

        # Current investment portfolio
        asset_types_input = input("Do you own stocks, bonds, or both? (stocks/bonds/both/none): ").lower()
        has_stocks = 'stocks' in asset_types_input or 'both' in asset_types_input
        has_bonds = 'bonds' in asset_types_input or 'both' in asset_types_input

        if has_stocks:
            enter_stock_details = input("Do you want to enter specific stock names and details? (yes/no): ").lower()
            if enter_stock_details == 'yes':
                num_stocks = int(input("How many different stocks do you want to enter? "))
                for i in range(num_stocks):
                    name = input(f"Enter ticker symbol for stock {i+1}: ")
                    quantity = float(input(f"Enter quantity for {name}: "))
                    buy_date = input(f"Enter purchase date for {name} (YYYY-MM-DD): ")
                    buy_price = float(input(f"Enter purchase price per share for {name} ({self.currency_symbol}): "))
                    self.stock_details[name] = {"quantity": quantity, "buy_date": buy_date, "buy_price": buy_price}
                    current_price = fetch_live_price(name)
                    if current_price is not None:
                        self.portfolio[name] = current_price * quantity
                        print(f"Fetched live price for {name}: {current_price:.2f} {self.currency_symbol}. Total value: {self.portfolio[name]:.2f} {self.currency_symbol}")
                    else:
                        print(f"Could not fetch live price for {name}. Using purchase price for value calculation.")
                        self.portfolio[name] = buy_price * quantity # Use buy price if fetching fails


        if has_bonds:
            enter_bond_details = input("Do you want to enter specific bond names and details? (yes/no): ").lower()
            if enter_bond_details == 'yes':
                num_bonds = int(input("How many different bonds do you want to enter? "))
                for i in range(num_bonds):
                    name = input(f"Enter ticker symbol for bond {i+1}: ")
                    quantity = float(input(f"Enter quantity for {name}: "))
                    buy_date = input(f"Enter purchase date for {name} (YYYY-MM-DD): ")
                    buy_price = float(input(f"Enter purchase price per bond for {name} ({self.currency_symbol}): "))
                    self.bond_details[name] = {"quantity": quantity, "buy_date": buy_date, "buy_price": buy_price}
                    # yfinance primarily focuses on stocks, bond data might be limited or require different handling.
                    current_price = fetch_live_price(name) # Still attempt to fetch
                    if current_price is not None:
                        self.portfolio[name] = current_price * quantity
                        print(f"Fetched live price for {name}: {current_price:.2f} {self.currency_symbol}. Total value: {self.portfolio[name]:.2f} {self.currency_symbol}")
                    else:
                        print(f"Could not fetch live price for {name}. Bond data through yfinance is often limited. Using purchase price for value calculation.")
                        self.portfolio[name] = buy_price * quantity # Use buy price if fetching fails


        # Other investments
        enter_other_investments = input("Do you have other investments (e.g., Crypto, Gold, Cash)? (yes/no): ").lower()
        if enter_other_investments == 'yes':
            num_other_investments = int(input("How many types of other investments do you have? "))
            for i in range(num_other_investments):
                name = input(f"Enter name for other investment {i+1} (e.g., Crypto, Gold, Cash): ")
                value = float(input(f"Enter current value for {name} ({self.currency_symbol}): "))
                self.other_investments_details[name] = value
                self.portfolio[f"other_{name}"] = value # Prefix with "other_" to distinguish from stocks/bonds


        # Financial goals
        self.goals = {
            "target_amount": float(input(f"Enter target amount for financial goals ({self.currency_symbol}): ")),
            "time_horizon": int(input("Enter time horizon for financial goals (in years): "))
        }

        # Risk tolerance
        self.risk_tolerance = input("Enter risk tolerance (conservative, moderate, aggressive): ")

        # Current financial situation
        self.financial_situation["debts"] = float(input(f"Enter total debts ({self.currency_symbol}): "))
        self.financial_situation["expenses"] = float(input(f"Enter monthly expenses ({self.currency_symbol}): "))
        self.financial_situation["salary"] = float(input(f"Enter your current annual salary ({self.currency_symbol}): "))

        # Other income
        self.financial_situation["other_income"]["interest"] = float(input(f"Enter annual interest income ({self.currency_symbol}): "))
        self.financial_situation["other_income"]["rent"] = float(input(f"Enter annual rental income ({self.currency_symbol}): "))
        self.financial_situation["annual_income"] = self.financial_situation["salary"] + self.financial_situation["other_income"]["interest"] + self.financial_situation["other_income"]["rent"]


        # Asset allocation (refined conditional input)
        print("\nEnter your desired asset allocation percentages (total should be 100%).")
        while True:
            try:
                stock_allocation = 0
                bond_allocation = 0
                other_allocation = 0

                if has_stocks:
                    stock_allocation = float(input("Enter desired stock allocation (%): "))
                    remaining_allocation = 100 - stock_allocation
                    print(f"Remaining allocation: {remaining_allocation:.2f}%")
                else:
                    remaining_allocation = 100

                if has_bonds:
                    bond_allocation = float(input(f"Enter desired bond allocation (%) (out of remaining {remaining_allocation:.2f}%): "))
                    remaining_allocation -= bond_allocation
                    print(f"Remaining allocation: {remaining_allocation:.2f}%")


                if self.other_investments_details: # Only ask if user has other investments
                     other_allocation = float(input(f"Enter desired other investments allocation (%) (out of remaining {remaining_allocation:.2f}%): "))
                     remaining_allocation -= other_allocation
                     print(f"Remaining allocation: {remaining_allocation:.2f}%")


                if abs(stock_allocation + bond_allocation + other_allocation - 100) < 1e-6: # Check if sum is close to 100
                    self.asset_allocation = {
                        "stocks": stock_allocation / 100,
                        "bonds": bond_allocation / 100,
                        "other": other_allocation / 100
                    }
                    break
                else:
                    print("Total allocation must add up to 100%. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")


        # Retirement goals and time horizon
        self.retirement_goals = {
            "target_amount": float(input(f"Enter target retirement amount ({self.currency_symbol}): ")),
            "time_horizon": int(input("Enter retirement time horizon (in years): "))
        }

        # Emergency fund amount and location
        self.emergency_fund["amount"] = float(input(f"Enter current emergency fund amount ({self.currency_symbol}): "))
        self.emergency_fund["location"] = input("Where is your emergency fund held (e.g., FD, savings account)? ")

        # Current tax situation and implications (basic)
        self.tax_situation["type"] = input("Are you a salaried person or a business person? (salaried/business): ").lower()
        if self.tax_situation["type"] == "salaried":
            self.tax_situation["regime"] = input("Which tax regime do you follow (old/new)? ").lower()
            self.tax_situation["implications"] = input("Enter any specific tax implications for salaried individuals: ")


        elif self.tax_situation["type"] == "business":
            self.tax_situation["implications"] = input("Enter any specific tax implications for business owners: ")

        else:
            self.tax_situation["implications"] = input("Enter any specific tax implications: ")


        # Insurance coverage and estate planning details (optional)
        self.insurance_coverage = input("Enter insurance coverage and estate planning details (optional): ")

        # Interest in ESG investments
        self.interest_in_esg = input("Enter interest in ESG investments (yes/no): ")

        # Other relevant factors
        self.other_factors = {
            "inflation_projections": float(input("Enter inflation projections (%): ")) / 100,
            "diversification": input("Enter desired diversification strategy: ")
        }


    def analyze_portfolio(self):
        # Calculate current portfolio value
        current_portfolio_value = sum(self.portfolio.values())

        # Calculate asset allocation alignment with goals
        # This calculation is a simplified representation and may need refinement
        calculated_allocation = {}
        total_portfolio_value = sum(self.portfolio.values())
        if total_portfolio_value > 0:
            stock_value = sum([value for key, value in self.portfolio.items() if key in self.stock_details])
            bond_value = sum([value for key, value in self.portfolio.items() if key in self.bond_details])
            other_value = sum([value for key, value in self.portfolio.items() if key.startswith("other_")])

            calculated_allocation["stocks"] = stock_value / total_portfolio_value
            calculated_allocation["bonds"] = bond_value / total_portfolio_value
            calculated_allocation["other"] = other_value / total_portfolio_value
        else:
            calculated_allocation = {"stocks": 0, "bonds": 0, "other": 0}


        asset_allocation_alignment = 0
        for asset_type in ['stocks', 'bonds', 'other']:
             if asset_type in self.asset_allocation and asset_type in calculated_allocation:
                 # A simple measure of alignment - can be made more sophisticated
                 alignment_percentage = 100 - abs(self.asset_allocation[asset_type] - calculated_allocation[asset_type]) * 100
                 asset_allocation_alignment += alignment_percentage * self.asset_allocation[asset_type] # Weighted average based on desired allocation


        # Calculate debt-to-income ratio
        # Handle potential division by zero if expenses are 0
        debt_to_income_ratio = self.financial_situation["debts"] / self.financial_situation["expenses"] if self.financial_situation["expenses"] != 0 else float('inf')

        # Calculate tax liability
        annual_income = self.financial_situation.get("annual_income", 0)
        tax_liability = self.calculate_tax(annual_income)


        return {
            "current_portfolio_value": current_portfolio_value,
            "asset_allocation_alignment": asset_allocation_alignment,
            "debt_to_income_ratio": debt_to_income_ratio,
            "tax_liability": tax_liability
        }

    def project_future_outcomes(self):
        # Project future portfolio value based on current investments and risk tolerance
        # This projection is a simplified representation and may need refinement
        # Consider adding more sophisticated growth rate calculations based on asset types and risk tolerance
        average_growth_rate = 0.05 # Placeholder - should be based on risk tolerance and asset allocation
        future_portfolio_value = sum(self.portfolio.values()) * (1 + average_growth_rate) ** self.goals["time_horizon"]

        # Project future debt repayment based on debt-to-income ratio
        # This projection might need refinement based on future debt planning strategy and interest rates
        # Assuming a fixed repayment rate for simplicity
        debt_repayment_rate = 0.05 # Placeholder
        future_debt_repayment = self.financial_situation["debts"] * (1 - debt_repayment_rate) ** self.goals["time_horizon"]


        return {
            "future_portfolio_value": future_portfolio_value,
            "future_debt_repayment": future_debt_repayment
        }

    def suggest_strategies(self):
        # Suggest asset allocation strategies to optimize portfolio
        # This suggestion is a simplified representation and may need refinement
        asset_allocation_strategy = "Review your current asset allocation and consider adjusting it to align with your desired percentages and risk tolerance."

        # Suggest debt management strategies
        debt_management_strategy = f"Focus on your future debt planning strategy: {self.future_debt_planning}. Consider prioritizing high-interest debts."

        investment_strategies = self.suggest_investment_strategies()


        return {
            "asset_allocation_strategy": asset_allocation_strategy,
            "debt_management_strategy": debt_management_strategy,
            "investment_strategies": investment_strategies
        }

    def calculate_tax(self, annual_income):
        """
        Calculates the tax liability based on annual income, country, and tax regime.
        Assumes 'salaried' tax type.
        """
        tax_liability = 0
        taxable_income = annual_income

        country = self.currency.lower()
        tax_type = self.tax_situation.get("type", "").lower()
        tax_regime = self.tax_situation.get("regime", "").lower()

        if tax_type != "salaried":
            return 0 # Tax calculation for other types is not implemented yet

        if country == "india":
            if tax_regime == "new":
                regime_info = tax_info["india"]["new_regime"]
                taxable_income = max(0, annual_income - regime_info.get("standard_deduction", 0))
                for slab in regime_info["slabs"]:
                    lower_bound, upper_bound = slab["income_range"]
                    rate = slab["rate"] / 100
                    if taxable_income > lower_bound:
                        taxable_in_slab = min(taxable_income, upper_bound) - lower_bound
                        tax_liability += taxable_in_slab * rate

                # Apply rebate under section 87A for new regime
                if annual_income <= regime_info.get("rebate_income_limit", 0):
                     tax_liability = max(0, tax_liability - regime_info.get("rebate_amount", 0))
                # Note: Marginal relief for income slightly above the rebate limit is not implemented here


            elif tax_regime == "old":
                # For simplicity, we'll use the 'below_60' slab for the old regime in this example
                # More sophisticated logic would be needed to determine the correct age-based slab and apply deductions
                regime_info = tax_info["india"]["old_regime"]["below_60"]
                taxable_income = max(0, annual_income - regime_info.get("standard_deduction", 0))
                # Note: Deductions like 80C would also apply here in a full implementation
                for slab in regime_info["slabs"]:
                    lower_bound, upper_bound = slab["income_range"]
                    rate = slab["rate"] / 100
                    if taxable_income > lower_bound:
                        taxable_in_slab = min(taxable_income, upper_bound) - lower_bound
                        tax_liability += taxable_in_slab * rate

                # Apply rebate under section 87A for old regime
                if annual_income <= regime_info.get("rebate_income_limit", 0):
                     tax_liability = max(0, tax_liability - regime_info.get("rebate_amount", 0))


            else:
                print(f"Unknown tax regime for India: {tax_regime}")
                return 0

        elif country == "usa":
            # For simplicity, we'll use the 'single' filing status for federal tax in this example
            # More sophisticated logic would be needed to determine the correct filing status and include state taxes
            regime_info = tax_info["usa"]["federal"]["single"]
            taxable_income = max(0, annual_income - regime_info.get("standard_deduction", 0))
            for slab in regime_info["slabs"]:
                lower_bound, upper_bound = slab["income_range"]
                rate = slab["rate"] / 100
                if taxable_income > lower_bound:
                    taxable_in_slab = min(taxable_income, upper_bound) - lower_bound
                    tax_liability += taxable_in_slab * rate

        else:
            print(f"Tax calculation not implemented for country: {country}")
            return 0

        # Add cess and surcharge if applicable (not implemented in this basic version)
        # For India, add Health and Education Cess @ 4%
        if country == "india":
            tax_liability *= 1.04


        return tax_liability

    def suggest_investment_strategies(self):
        """
        Suggests investment strategies based on user goals, risk tolerance, and financial situation.
        """
        strategies = []

        # Strategy based on goals and time horizon
        if self.goals.get("time_horizon", 0) > 10:
            strategies.append("Consider Systematic Investment Plans (SIPs) for your long-term goals to benefit from rupee cost averaging.")
        elif self.goals.get("time_horizon", 0) > 5:
            strategies.append("For your medium-term goals, a balanced approach with a mix of equity and debt may be suitable.")
        else:
            strategies.append("For short-term goals, focus on low-risk investments like fixed deposits or short-term debt funds.")

        # Strategy based on risk tolerance
        if self.risk_tolerance.lower() == "aggressive":
            strategies.append("Given your aggressive risk tolerance, consider a higher allocation to equity and potentially explore growth stocks or sector-specific funds.")
        elif self.risk_tolerance.lower() == "moderate":
            strategies.append("With a moderate risk tolerance, a diversified portfolio across different asset classes is recommended.")
        else: # Conservative
            strategies.append("For a conservative risk tolerance, prioritize capital preservation with a focus on debt instruments and stable assets.")

        # Strategy based on financial situation (debt)
        if self.financial_situation.get("debts", 0) > 0 and self.financial_situation["debts"] > self.financial_situation.get("annual_income", 0) * 0.5: # Example threshold
            strategies.append(f"Your current debt level ({self.financial_situation['debts']:.2f} {self.currency_symbol}) is relatively high. Prioritize reducing high-interest debts before significantly increasing investments.")
        elif self.financial_situation.get("debts", 0) > 0:
             strategies.append(f"Manage your existing debts ({self.financial_situation['debts']:.2f} {self.currency_symbol}) effectively. Consider strategies like debt consolidation or accelerated payments.")
        else:
            strategies.append("Your debt level is manageable, allowing for more flexibility in investment planning.")


        # Strategy based on emergency fund
        # A common rule of thumb is 3-6 months of expenses
        recommended_emergency_fund = self.financial_situation.get("expenses", 0) * 6
        if self.emergency_fund.get("amount", 0) < recommended_emergency_fund:
             strategies.append(f"Your emergency fund ({self.emergency_fund.get('amount', 0):.2f} {self.currency_symbol}) is currently below the recommended level of {recommended_emergency_fund:.2f} {self.currency_symbol} (6 months of expenses). Building a sufficient emergency fund should be a priority.")
        else:
            strategies.append("Your emergency fund is well-funded, providing a good financial safety net.")


        # Strategy based on tax situation
        if self.tax_situation.get("type", "").lower() == "salaried" and self.tax_situation.get("regime", "").lower() == "old":
             strategies.append("If you are following the old tax regime, explore tax-saving investment options like PPF, ELSS, and NPS.")
        elif self.tax_situation.get("type", "").lower() == "salaried" and self.tax_situation.get("regime", "").lower() == "new":
             strategies.append("If you are following the new tax regime, tax-saving investments may not provide significant benefits. Focus on investments aligned with your financial goals and risk tolerance.")

        # Strategy based on ESG interest
        if self.interest_in_esg.lower() == "yes":
             strategies.append("Explore ESG (Environmental, Social, and Governance) investment options that align with your values.")


        return strategies


    def provide_recommendations(self):
        analysis_results = self.analyze_portfolio()
        future_outcomes = self.project_future_outcomes()
        strategies = self.suggest_strategies()

        print(f"\n--- Financial Analysis Summary ({self.currency.capitalize()}) ---")
        print(f"Current portfolio value: {analysis_results['current_portfolio_value']:.2f} {self.currency_symbol}")
        print(f"Asset allocation alignment with goals: {analysis_results['asset_allocation_alignment']:.2f}%")
        print(f"Debt-to-income ratio: {analysis_results['debt_to_income_ratio']:.2f}")
        print(f"Estimated Annual Tax Liability: {analysis_results['tax_liability']:.2f} {self.currency_symbol}")

        print("\n--- Future Projections ---")
        print(f"Projected future portfolio value: {future_outcomes['future_portfolio_value']:.2f} {self.currency_symbol}")
        print(f"Projected future debt repayment: {future_outcomes['future_debt_repayment']:.2f} {self.currency_symbol}")

        print("\n--- Recommendations ---")
        print(f"Asset Allocation Strategy: {strategies['asset_allocation_strategy']}")
        print(f"Debt Management Strategy: {strategies['debt_management_strategy']}")
        print(f"Emergency Fund Location: Your emergency fund is in a {self.emergency_fund['location']} and currently stands at {self.emergency_fund['amount']:.2f} {self.currency_symbol}.")
        print(f"Tax Situation: You are a {self.tax_situation['type']} person.")
        if self.tax_situation['type'] == 'salaried':
            print(f"  Tax Regime: {self.tax_situation['regime'].capitalize()}")
        print(f"  Tax Implications: {self.tax_situation['implications']}")

        print("\n--- Suggested Investment Strategies ---")
        for strategy in strategies['investment_strategies']:
            print(f"- {strategy}")


        print("\n--- Suggested Action Plan ---")
        print("1. Review and potentially adjust your asset allocation based on your goals and risk tolerance.")
        print(f"2. Implement your future debt planning strategy: {self.future_debt_planning}")
        if self.interest_in_esg.lower() == 'yes':
             print("3. Explore ESG investment options if they align with your values and financial strategy.")
        print("4. Regularly review and update your financial plan as your situation and goals change.")
        print("5. Consider consulting a financial advisor for personalized guidance.")



def main():
    analyzer = FinancialAnalyzer()
    analyzer.collect_user_inputs()
    analyzer.analyze_portfolio()
    analyzer.project_future_outcomes()
    analyzer.suggest_strategies()
    analyzer.provide_recommendations()

if __name__ == "__main__":
    main()
```
