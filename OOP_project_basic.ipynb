```python
import nlpcloud
class NlpApp:

    def __init__(self):
        self.__database = {}
        self.__first_menu()

    def __first_menu(self):
        first_input = int(input("""

Hello! Do you want to login or sign up? It's up to you.

1. Log in
2. Sign up
3. Exit

"""))

        if first_input == 1:
            self.__register()
        elif first_input == 2:
            self.__login()
        else:
            exit()
    def __second_menu(self):
        second_input = int(input("""

Hello! Do you want to login or sign up? It's up to you.

1. NER
2. Language detection
3. sentiment
4. logout

"""))

        if second_input == 1:
            self.__ner()
        elif second_input == 2:
            self.__detection()
        elif second_input ==3:
          self.__sentiment
        else:
            exit()

    def __sentiment(self):
      para = input('ente the paragraph :')
      #search_term = input('what is the would you want to check : ')
      client = nlpcloud.Client("finetuned-llama-3-70b", "6dd5efb2024d9e59712d220c9df00644607e7f71", gpu=True)
      response = client.sentiment(para, target="NLP Cloud") # Assign the result to a variable

      L = []
      for i in response['scored_labels']:
        L.append(i['score'])

      index = sorted(list(enumerate(L)), key = lambda x:x[1], reverse=True)[0][0]

      print(response['scored_labels'][index]['label'])


      print(response)
      self.__second_menu

    def __ner(self):
      para = input('ente the paragraph :')
      search_term = input('what is the would you want to check : ')
      client = nlpcloud.Client("finetuned-llama-3-70b", "6dd5efb2024d9e59712d220c9df00644607e7f71", gpu=True)
      response = client.entities(para, searched_entity= search_term) # Assign the result to a variable

      print(response)
      self.__second_menu



    def __register(self):
        name = input('Enter your name: ')
        email = input('Enter your email: ')
        password = input('Enter your password: ')

        if email in self.__database:
            print("Account already exists.")
        else:
            self.__database[email] = [name, password]
            print("Registered successfully!\n")
            self.__first_menu()

    def __login(self):
        email = input('Enter your email: ')
        password = input('Enter your password: ')

        if email in self.__database and password == self.__database[email][1]:
            print('Login successful')
            self.__second_menu()
        else:
            print('Incorrect email or password. Try again.\n')
            self.__login()


a = NlpApp()
```
